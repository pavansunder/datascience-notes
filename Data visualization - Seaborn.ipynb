# Distribution plots

### Distribution plot - distribution of continuous data

```python
sb.distplot(tips['total_bill'], bins =[0, 10, 20, 30, 40, 50, 60])
```

### Histogram plot

```python
sb.distplot(tips['total_bill'], bins =[0, 10, 20, 30, 40, 50, 60], kde=False)
```

### kde plot

```python
sb.kdeplot(tips['total_bill'])
```

### Rug plot - shows distribution by drawing ticks

```python
sb.rugplot(tips['total_bill'], height=0.1)
```

```python
# rugplots are used in combination with other plots..For example

sb.kdeplot(tips['total_bill'])
sb.rugplot(tips['total_bill'], height=0.05)
```

### Pair plot

```python
sb.pairplot(iris, hue='species')
```

### Joint plot

```python
sb.jointplot(x='petal_length',y='petal_width', data=iris, kind='reg')
```

### styling

#### available styles : darkgrid, whitegrid, dark, white, ticks

```python
plt.figure(figsize=(15,4))
sb.set_style('darkgrid')
#sb.set_context(context='paper', font_scale=1.8)
sb.distplot(tips['total_bill'], bins =[0, 10, 20, 30, 40, 50, 60])

plt.title('total bill distribution plot')
```

# Matrix plots

### Heat maps

```python
# iris dataset
plt.figure(figsize=(8,6))
iris_corr = iris.corr()
sb.heatmap(iris_corr, annot=True, cmap='coolwarm')
```

```python
# flights dataset

flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')
flights_pivot
```

```python
plt.figure(figsize=(8,6))
sb.heatmap(flights_pivot, cmap='Blues', linecolor='white', linewidths=0.5)
```

### Cluster map

```python
# iris dataset

#species=iris.pop('species')
sb.clustermap(iris,
             row_cluster=False,
             figsize=(7,5),
             cbar_pos=(0.1, .2, .02, .2),
             cmap='mako',
             #metric='correlation',
             vmin=0,
             vmax=10)
```

```python
#flights dataset

sb.clustermap(flights_pivot,
              #col_cluster=False,
              cmap='Blues',
              figsize=(7,5),
              cbar_pos=(0.1, .2, .02, .2),
              standard_scale=1)
```

### Pair grid

```python
iris_g = sb.PairGrid(iris, hue='species')
iris_g.map_diag(plt.hist)
#iris_g.map_offdiag(sb.scatterplot)
iris_g.map_upper(sb.scatterplot)
iris_g.map_lower(sb.kdeplot)
iris_g.add_legend()
```

```python
# custom pair grid
iris_g = sb.PairGrid(iris, hue='species',
                     x_vars=['sepal_length', 'sepal_width'],
                     y_vars=['petal_length', 'petal_width'])
iris_g.map(sb.scatterplot)
iris_g.add_legend()
```

### Facet grid

```python
tips_fg = sb.FacetGrid(tips, row='smoker', col='time')
tips_fg.map(plt.scatter, 'total_bill', 'tip')
```

```python
tips_fg = sb.FacetGrid(tips, col='time', hue='smoker', height=4, aspect=1.5, palette='Set1')
tips_fg.map(plt.scatter, 'total_bill', 'tip', edgecolor='w')
tips_fg.add_legend()
```

```python
kw=dict(s=50, linewidth=0.5, edgecolor='b')
tips_fg = sb.FacetGrid(tips, col='sex', hue='smoker', 
                       height=4, 
                       aspect=1.5, 
                       palette='Set1',
                       hue_kws={'marker':['^','v']})
tips_fg.map(plt.scatter, 'total_bill', 'tip', **kw)
tips_fg.add_legend()
```

```python
flights_fg = sb.FacetGrid(flights, 
                       col='year',
                       col_wrap=4,
                       height=4, 
                       aspect=1, 
                       palette='Set1')
flights_fg.map_dataframe(sb.barplot, 'passengers', 'month')
```
