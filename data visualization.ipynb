# *data visualization*

```python
import matplotlib.pyplot as plt # or from matplotlib import pyplot as plt
```

## *Line plot*

```python
x=[1,2,3]
y=[4,5,2]
plt.plot(x,y)
plt.title('New graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
x1=[1,2,3]
y1=[4,5,2]
x2=[6,7,8]
y2=[9,10,7]
plt.plot(x1,y1,'green',label='first',linewidth=1)
plt.plot(x2,y2,'m',label='second',linewidth=2) # we can use short form also to specify colour
plt.title('Style graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.grid(True,color='k')
plt.show()
```

## *Bar plot*

#### applicable for categorical data

```python
x1=[1,3,5,7]
y1=[5,2,6,6]

x2=[2,4,6,8]
y2=[6,3,5,2]

plt.bar(x1,y1,label='First')
plt.bar(x2,y2,label='Second',color='green')
plt.title('Bar plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.show()
```

## *Histogram*

#### applicable for continuous data

```python
sales_values=[12,10,5,7,0,14,15,24,10,11,15,30,18,11,10,18,30]
bins=[0,5,10,15,20,25,30]
plt.hist(sales_values,bins,histtype='bar',rwidth=0.6,color='green') # hist type: we can also give barstacked,step,stepfilled
plt.xlabel('X axis')
plt.ylabel('y axis')
```

## *Scatter plot*

#### important to understand relationships

```python
x=[1,3,4,8,5,6,7]
y=[3,5,6,7,8,9,3]
plt.scatter(x,y)
plt.title('scatter plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

## *Stack plot*

#### when we have values changing over days or weeks based on range

```python
days=[1,2,3,4,5]
sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]
plt.stackplot(days,sleeping,eating,working,playing,colors=['m','c','r','k'],labels=['sleeping','eating','working','playing'])
plt.title('daily routine')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.show()
```

## *Pie chart*

#### when we have to get the proportion of different categories

```python
labels=['python','c++','ruby','java']
sizes=[215,130,254,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.2f%%',shadow=True,startangle=90)
plt.axis('equal')
plt.show()
```

### seaborn vs matplotlib

```python
# with matplotlib
import numpy as np
from matplotlib import pyplot as plt

def sinplot(flip=1):
    x=np.linspace(0,14,100)
    for i in range(1,5):
        plt.plot(x,np.sin(x+i*0.5)*(7-i)*flip)
        
sinplot()
plt.show()
```

```python
# with seaborn
sb.set()
sinplot()
plt.show()
```

### colour pallete

### Types of colour pallete:
#### 1.qualitative
#### 2.sequential
#### 3.diverging

```python
# qualitative:
current_palette=sb.color_palette()
sb.palplot(current_palette)
plt.show()
```

```python
# sequential:
current_palette=sb.color_palette("Greens")
sb.palplot(current_palette)
plt.show()
```

```python
# diverging:
current_palette=sb.color_palette('BrBG',7)
sb.palplot(current_palette)
plt.show()
```

```python
# using coloue palette (few are : husl,summer,winter,spring,autumn)

sb.set_palette('winter')
sinplot()
plt.show()
```
