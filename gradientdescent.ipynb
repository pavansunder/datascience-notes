```python
import pandas as pd
import numpy as np
import seaborn as sb
% matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(20.0,20.0)
data=pd.read_csv('E:/important notes and lectures/python/INVENTATEQ/datascience/datasets/headbrain.csv)
print(data.shape)
data.head()
```

```python
#collecting X and Y
X=data[0].values
Y=data[1].values
n=len(X)

#plotting
plt.plot(X,Y)
plt.title('Signal')
plt.xlabel('Duration')
plt.ylabel('Normalized Amplitude')
plt.show()
```

```python

#getting regression lines for random values b1

#set b0 = 0 (any random value, here i have set to 0)
b0=0
# Generate random weights
b1=list(np.arange(-0.3,0.8,0.01))

Y_pred=[]
for slope in b1:
    Y_pred.append([b0+slope*x for x in X])

#plot regression lines
for line in Y_pred:
    plt.plot(X,line)
```

```python
def RootMeanSquared(k,Y):
    MSE=0
    n=len(Y)
    for i in range(n):
        MSE+=(Y[i]-Y_pred[k][i])**2/n
    RMSE=np.sqrt(MSE)
    return RMSE    
```

```python
MSE_list=[RootMeanSquared(k,Y) for k in range(len(b1))]
```

```python
plt.plot(b1,MSE_list)
```

```python
minima=min(MSE_list)
idx=MSE_list.index(minima)
print('Minima:',minima,'index:',idx)
```

```python
plt.plot(X,Y_pred[idx])

sb.scatterplot(X,Y)
plt.title('Correlation between variables')
plt.xlabel('Head size in cm^3')
plt.ylabel('Brain weight in grams')
plt.show()
```

```python
ssr=0
sst=0
n=len(X)
mean_Y=np.mean(Y)

for i in range(n):
    ssr+=(Y[i]-Y_pred[idx][i])**2
    sst+=(Y[i]-mean_Y)**2
    
r_sqr=1-(ssr/sst)
print('r2_score:',r_sqr)
```
